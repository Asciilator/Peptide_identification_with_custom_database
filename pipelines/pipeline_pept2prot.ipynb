```python
# Define the wrangle_peptides function
import re

def wrangle_peptides(sequence: str, ptm_filter: bool=True, li_swap: bool=True) -> str:
    """Process peptide sequences by removing post-translational modifications
    and/or equating Leucine and Isoleucine amino acids.

    Args:
        sequence (str): Peptide sequence string
        ptm_filter (bool, optional): Remove PTMs from sequence. Defaults to True.
        li_swap (bool, optional): Equate Leucine (L) and Isoleucine (I). Defaults to True.

    Returns:
        str: Processed sequence string
    """
    if ptm_filter:
        sequence = "".join(re.findall(r"[A-Z]+", sequence))
    if li_swap:
        sequence = sequence.replace("L", "I")
    return sequence

# Apply wrangle_peptides function to the filtered data
filtered_df["Cleaned Sequence"] = filtered_df["Peptide"].apply(lambda x: wrangle_peptides(x))

print("Shape of the filtered DataFrame after adding the new column:")
display(filtered_df.shape)  # Check the shape of the DataFrame after adding the new column

# Display rows 11 to 21 (Python is 0-indexed, so need 10:21)
print("Cleaned Peptides (Rows 11â€“21):")
display(filtered_df[["Peptide", "Cleaned Sequence"]].iloc[10:21])
```

```python
# A disadvantage of using pept2prot like this is that takes way longer than doing the same with pept2lca
import pandas as pd
import requests
import time

def fetch_request(url: str, retries: int = 3, delay: int = 5) -> requests.Response:
    for attempt in range(retries):
        req_get = requests.get(url)
        if req_get.status_code == 200:
            return req_get
        print(f"Request failed (status {req_get.status_code}). Retrying {attempt+1}/{retries}...")
        time.sleep(delay)
    raise RuntimeError(f"Request failed after {retries} retries: {req_get.status_code}")

def request_unipept_pept2prot(pept_df: pd.DataFrame, seq_col: str) -> pd.DataFrame:
    base_url = "http://api.unipept.ugent.be/api/v1/pept2prot.json?equate_il=true"
    batch_size = 100
    peptides = pept_df[seq_col].dropna().drop_duplicates().tolist()
    all_rows = []

    for i in range(0, len(peptides), batch_size):
        batch = peptides[i:i+batch_size]
        query = "&".join([f"input[]={p}" for p in batch])
        url = f"{base_url}&{query}"
        print(f"ğŸ” Sending batch {i//batch_size + 1}/{(len(peptides) - 1)//batch_size + 1}...")
        
        try:
            response = fetch_request(url).json()
            for record in response:
                all_rows.append({
                    "Peptide": record.get("peptide", "Unknown"),
                    "UniProt ID": record.get("uniprot_id", "Unknown"),
                    "Protein Name": record.get("protein_name", "Unknown"),
                    "Taxon ID": record.get("taxon_id", "Unknown"),
                    "Protein Sequence": record.get("protein", "Unknown")
                })
        except Exception as e:
            print(f"âš ï¸ Failed batch starting with {batch[0]}: {e}")
            continue

    df = pd.DataFrame(all_rows)
    print(f"âœ… Done. Total mappings: {len(df)}")
    return df
```

```python
# Run it on your cleaned peptides
pept2prot_df = request_unipept_pept2prot(filtered_df, seq_col="Cleaned Sequence")

# Show first rows
display(pept2prot_df.head())
```

```python
def filter_by_peptide_support(df, min_peptides=5):
    counts = df.groupby("UniProt ID")["Peptide"].nunique().reset_index()
    counts = counts[counts["Peptide"] >= min_peptides]
    filtered = df[df["UniProt ID"].isin(counts["UniProt ID"])]
    print(f"âœ… Retained {len(filtered)} mappings for {len(counts)} proteins supported by â‰¥ {min_peptides} peptides.")
    return filtered
```

```python
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio import SeqIO
import os

# Define output path in your project folder
fasta_output_path = "unique_filtered_pept2prot.fasta"

# Remove duplicate proteins by UniProt ID + sequence
unique_proteins = pept2prot_df_filtered.drop_duplicates(subset=["UniProt ID", "Protein Sequence"])

# Convert to SeqRecord format
records = []
for _, row in unique_proteins.iterrows():
    header = f"{row['UniProt ID']}|{row['Protein Name']}|taxid:{row['Taxon ID']}"
    seq = Seq(row["Protein Sequence"])
    record = SeqRecord(seq, id=header, description="")
    records.append(record)

# Write to FASTA
SeqIO.write(records, fasta_output_path, "fasta")
print(f"âœ… Exported {len(records)} unique proteins to: {fasta_output_path}")
```

```python
from ete3 import NCBITaxa
from pathlib import Path
from Bio import SeqIO
import re

# === Setup paths ===
pept2prot_fasta_path = base_path / "pipelines" / "unique_filtered_pept2prot.fasta"
output_fasta_path = pept2prot_fasta_path.parent / "unique_filtered_pept2prot_bacteria.fasta"

# === Initialize NCBI Taxonomy
ncbi = NCBITaxa()

# === Helper: Extract taxid from header formatted as 'taxid:xxxxx'
def extract_taxid(header):
    match = re.search(r"taxid:(\d+)", header)
    return match.group(1) if match else None

# === Load and filter records for Bacteria only
bacterial_records = []
for record in SeqIO.parse(pept2prot_fasta_path, "fasta"):
    taxid = extract_taxid(record.description)
    if taxid:
        try:
            lineage = ncbi.get_lineage(int(taxid))
            names = ncbi.get_taxid_translator(lineage)
            if "Bacteria" in names.values():
                bacterial_records.append(record)
        except Exception:
            continue

# === Write output
SeqIO.write(bacterial_records, output_fasta_path, "fasta")

# === Summary
print(f"âœ… Bacterial proteins retained: {len(bacterial_records)}")
print(f"ğŸ“ Output saved to: {output_fasta_path}")
```

```python
from Bio import SeqIO
import pandas as pd
import re
from pathlib import Path
from ete3 import NCBITaxa

# === Setup paths ===
fasta_path = base_path / "pipelines" / "unique_filtered_pept2prot_bacteria.fasta"
metagenomics_path = base_path / "db_results_analysis" / "unique_taxa_in_metagendb.csv"

# === Initialize NCBI Taxonomy
ncbi = NCBITaxa()

# === Step 1: Extract taxon IDs from FASTA and translate to names
fasta_taxa = set()

for record in SeqIO.parse(fasta_path, "fasta"):
    match = re.search(r"taxid:(\d+)", record.description)
    if match:
        taxid = int(match.group(1))
        try:
            name = ncbi.get_taxid_translator([taxid])[taxid]
            fasta_taxa.add(name.strip())
        except:
            continue

# === Step 2: Load taxa from metagenomics DB
df_meta = pd.read_csv(metagenomics_path)
df_meta.columns = df_meta.columns.str.strip()
meta_taxa = set(df_meta["taxon_name"].dropna().str.strip())

# === Step 3: Compare
overlap = fasta_taxa & meta_taxa
only_in_fasta = fasta_taxa - meta_taxa
only_in_meta = meta_taxa - fasta_taxa

# === Step 4: Report
total_fasta = len(fasta_taxa)
total_meta = len(meta_taxa)
overlap_count = len(overlap)
percent_overlap = (overlap_count / total_fasta) * 100 if total_fasta > 0 else 0

# === Step 5: Print
print(f"ğŸ§¬ Total unique taxa in pept2prot (bacteria): {total_fasta}")
print(f"ğŸ§¬ Total unique taxa in Metagenomics DB: {total_meta}")
print(f"âœ… Overlapping taxa: {overlap_count}")
print(f"âŒ pept2prot taxa not in Metagenomics: {len(only_in_fasta)}")
print(f"âŒ Metagenomics taxa not in pept2prot: {len(only_in_meta)}")
print(f"ğŸ“Š Percent of pept2prot taxa overlapping with Metagenomics: {percent_overlap:.2f}%")
```

```python
# This script combines overlapping proteins between the pept2prot and pept2taxa fasta databases into a single database
from Bio import SeqIO
from pathlib import Path
import pandas as pd

# Define input FASTA files
pept2prot_path = Path("uniprot_matched_proteins.fasta")
pept2taxa_path = Path("fasta_proteomes_proteomes_api/session_strain_level_2025-05-14_16-06-54\pept2taxa_strain_200_db.fasta")

# Define output path
combined_dir = Path("fasta_proteomes/combined_databases")
combined_dir.mkdir(parents=True, exist_ok=True)
output_path = combined_dir / "pept2taxa_strain200_pept2prot_combined.fasta"

# Read both FASTA files into dictionaries
pept2prot_records = SeqIO.to_dict(SeqIO.parse(pept2prot_path, "fasta"))
strain_records = SeqIO.to_dict(SeqIO.parse(pept2taxa_path, "fasta"))

# Find overlapping protein sequences (based on sequence string)
strain_sequences = {str(r.seq) for r in strain_records.values()}
overlap = {
    seq_id: rec for seq_id, rec in pept2prot_records.items()
    if str(rec.seq) in strain_sequences
}

# Write overlapping proteins to new FASTA
SeqIO.write(overlap.values(), output_path, "fasta")

# Show summary
print(f"âœ… Total in pept2prot: {len(pept2prot_records)}")
print(f"âœ… Total in strain proteomes: {len(strain_records)}")
print(f"âœ… Overlapping proteins: {len(overlap)}")
print(f"ğŸ“„ Overlapping proteins written to: {output_path}")
```
