```python
# Define the wrangle_peptides function
import re

def wrangle_peptides(sequence: str, ptm_filter: bool=True, li_swap: bool=True) -> str:
    """Process peptide sequences by removing post-translational modifications
    and/or equating Leucine and Isoleucine amino acids.

    Args:
        sequence (str): Peptide sequence string
        ptm_filter (bool, optional): Remove PTMs from sequence. Defaults to True.
        li_swap (bool, optional): Equate Leucine (L) and Isoleucine (I). Defaults to True.

    Returns:
        str: Processed sequence string
    """
    if ptm_filter:
        sequence = "".join(re.findall(r"[A-Z]+", sequence))
    if li_swap:
        sequence = sequence.replace("L", "I")
    return sequence

# Apply wrangle_peptides function to the filtered data
filtered_df["Cleaned Sequence"] = filtered_df["Peptide"].apply(lambda x: wrangle_peptides(x))

print("Shape of the filtered DataFrame after adding the new column:")
display(filtered_df.shape)  # Check the shape of the DataFrame after adding the new column

# Display rows 11 to 21 (Python is 0-indexed, so need 10:21)
print("Cleaned Peptides (Rows 11‚Äì21):")
display(filtered_df[["Peptide", "Cleaned Sequence"]].iloc[10:21])
```

```python
import requests
import pandas as pd
import time
from io import StringIO

# === CONFIG ===
CHUNK_SIZE = 500  # You can increase this to ~9500 once you're sure it's stable
SLEEP_BETWEEN_JOBS = 5  # seconds between submissions

# === STEP 1: Extract unique accessions from diamond_df ===
accessions = diamond_df["accession"].dropna().unique().tolist()
print(f"üî¢ Total unique accessions: {len(accessions)}")

# === UniProt Batch Mapping Helper Functions ===
def submit_uniprot_mapping(accessions_chunk):
    url = "https://rest.uniprot.org/idmapping/run"
    headers = {"Content-Type": "application/x-www-form-urlencoded"}
    data = {
        "from": "UniProtKB_AC-ID",
        "to": "UniProtKB",
        "ids": ",".join(accessions_chunk)
    }

    response = requests.post(url, data=data, headers=headers)
    response.raise_for_status()
    return response.json()["jobId"]

def wait_for_completion(job_id):
    url = f"https://rest.uniprot.org/idmapping/status/{job_id}"
    while True:
        response = requests.get(url)
        response.raise_for_status()
        status = response.json()

        if status.get("jobStatus", "").upper() == "RUNNING":
            print("‚è≥ Still processing...")
            time.sleep(3)
        elif "redirectURL" in status or "results" in status or status.get("jobStatus", "").upper() == "FINISHED":
            print("‚úÖ Job completed.")
            break
        else:
            print("üîÅ Status unclear, retrying...")
            time.sleep(3)

def download_results(job_id):
    print("‚¨áÔ∏è Downloading all paginated results...")
    base_url = f"https://rest.uniprot.org/idmapping/uniprotkb/results/{job_id}"
    headers = {"Accept": "text/tab-separated-values"}

    all_chunks = []
    next_url = base_url + "?format=tsv&fields=accession,organism_name&size=500"

    while next_url:
        response = requests.get(next_url, headers=headers)
        response.raise_for_status()
        chunk = pd.read_csv(StringIO(response.text), sep="\t")
        all_chunks.append(chunk)

        # Check if there's a next page in the headers
        link_header = response.headers.get("Link", "")
        if 'rel="next"' in link_header:
            parts = link_header.split(",")
            next_links = [p for p in parts if 'rel="next"' in p]
            if next_links:
                next_url = next_links[0].split(";")[0].strip("<> ")
            else:
                next_url = None
        else:
            next_url = None

    full_df = pd.concat(all_chunks, ignore_index=True)
    print(f"‚úÖ Downloaded total {len(full_df)} mapped accessions.")
    return full_df

# === STEP 2: Chunk accessions and map in batches ===
all_mappings = []

for i in range(0, len(accessions), CHUNK_SIZE):
    chunk = accessions[i:i+CHUNK_SIZE]
    print(f"\nüöÄ Submitting chunk {i//CHUNK_SIZE + 1}/{(len(accessions)//CHUNK_SIZE)+1} (size: {len(chunk)})")

    try:
        job_id = submit_uniprot_mapping(chunk)
        print(f"üÜî Job ID: {job_id}")
        wait_for_completion(job_id)
        result_df = download_results(job_id)
        all_mappings.append(result_df)
        print(f"‚úÖ Mapped {len(result_df)} accessions.")
    except Exception as e:
        print(f"‚ùå Error during chunk {i//CHUNK_SIZE + 1}: {e}")

    time.sleep(SLEEP_BETWEEN_JOBS)

# === STEP 3: Merge all results and map to diamond_df ===
if all_mappings:
    combined_df = pd.concat(all_mappings, ignore_index=True)
    print(f"\nüéØ Total mapped entries: {len(combined_df)}")

    # Create mapping dictionary (From ‚Üí Organism)
    accession_to_org = dict(zip(combined_df["From"], combined_df["Organism"]))

    # Map back to diamond_df
    diamond_df["organism"] = diamond_df["accession"].map(accession_to_org)

    # Save updated DataFrame
    output_path = "Diamond_alignments/diamond_df_with_organisms.csv"
    diamond_df.to_csv(output_path, index=False)
    print(f"üíæ Saved updated diamond_df with organisms to: {output_path}")
else:
    print("‚ö†Ô∏è No successful mappings. Check API or chunk errors.")
```

```python
# This script fetches the proteome IDs for the (filtered) organisms
import pandas as pd
import os
import requests
from datetime import datetime
from tqdm import tqdm

# === 1. Setup output path ===
session_time = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
output_dir = f"Diamond_alignments/session_diamond_align_{session_time}"
os.makedirs(output_dir, exist_ok=True)

proteome_csv_path = os.path.join(output_dir, "diamond_proteome_matches.csv")
combined_fasta_path = os.path.join(output_dir, "diamond_aligned_from_organisms.fasta")

# === 2. Prepare a DataFrame with organism names and dummy taxon IDs ===
# Load filtered organism-matching results
filtered_df = pd.read_csv("Diamond_alignments/diamond_df_filtered_confident.csv")

# Get unique, confident organism names
organism_list = filtered_df["organism"].dropna().unique().tolist()

taxa_df = pd.DataFrame({
    "Taxon ID": [0]*len(organism_list),  # Dummy taxon ID
    "Taxon Name": organism_list
})

# === 3. Proteome search function (same as pept2taxa pipeline) ===
def search_best_proteomes(taxa_df, name_col="Taxon Name", id_col="Taxon ID"):
    found = []
    not_found = []

    for _, row in tqdm(taxa_df.iterrows(), total=len(taxa_df)):
        taxon_id = row[id_col]
        taxon_name = row[name_col]

        url = "https://rest.uniprot.org/proteomes/search"
        used_query = "name"

        # Step 1: Try by organism name
        params = {"query": taxon_name, "format": "json", "size": 100}
        try:
            r = requests.get(url, params=params)
            r.raise_for_status()
            results = r.json().get("results", [])
        except:
            results = []

        # Step 2: Fallback to taxon ID (only if it's real)
        if not results and taxon_id:
            params = {"query": f"taxonomy_id:{taxon_id}", "format": "json", "size": 100}
            try:
                r = requests.get(url, params=params)
                r.raise_for_status()
                results = r.json().get("results", [])
                used_query = "taxid"
            except:
                results = []

        # Step 3: Handle failure
        if not results:
            not_found.append((taxon_id, taxon_name))
            print(f"‚ùå No proteome found for: {taxon_name} (Taxon ID: {taxon_id})")
            continue

        # Step 4: Pick best proteome
        reference = [r for r in results if r.get("proteomeType") == "Reference"]
        selected = reference[0] if reference else max(
            results, key=lambda r: r.get("busco", {}).get("complete", 0)
        )

        proteome_id = selected.get("id", None)
        is_ref = selected.get("proteomeType") == "Reference"

        print(f"‚úÖ {taxon_name} ‚Üí Proteome ID: {proteome_id} (Reference: {is_ref}, via: {used_query})")

        found.append((taxon_id, taxon_name, proteome_id, is_ref))

    print(f"\nüéØ Proteomes found for {len(found)} of {len(taxa_df)} organisms.")
    if not_found:
        print(f"\n‚ö†Ô∏è No proteomes found for {len(not_found)} organisms.")
    return found
```

```python
# The script will read the CSV file containing the proteome IDs, download the corresponding FASTA files, and save them into a combined FASTA file.import requests
# So only use this if you already have the proteome IDs in a CSV file.
# The script uses a download loop using Python‚Äôs ThreadPoolExecutor, which lets you download many proteomes in parallel, which is way faster than downloading them one by one.
import requests
import pandas as pd
import os
from concurrent.futures import ThreadPoolExecutor, as_completed

# === Function to download proteome FASTA ===
def download_fasta_entry(row):
    upid = row["Proteome ID (UPID)"]
    name = row["Taxon Name"]
    url = f"https://rest.uniprot.org/uniprotkb/stream?compressed=false&format=fasta&query=(proteome:{upid})"

    try:
        response = requests.get(url, timeout=60)
        if response.ok and response.text.startswith(">"):
            count = response.text.count(">")
            print(f"‚úÖ {upid} ({name}): {count} proteins")
            return response.text
        else:
            print(f"‚ùå Failed to download {upid} ({name})")
    except Exception as e:
        print(f"‚ùå Error for {upid}: {e}")
    return None

# === Read proteome list ===
proteome_csv_path = r"C:\Users\Yusuf\OneDrive\LST\Derde_jaar\Y3Q4\Pipeline\Diamond_alignments\session_diamond_align_2025-05-15_01-12-37\diamond_proteome_matches.csv"
df = pd.read_csv(proteome_csv_path)
valid_df = df[df["Proteome ID (UPID)"].notna()].reset_index(drop=True)

# === Set output FASTA path ===
csv_basename = os.path.basename(proteome_csv_path).replace(".csv", "")
combined_fasta_path = os.path.join(os.path.dirname(proteome_csv_path), f"{csv_basename}_combined.fasta")

# === Download in parallel ===
total_proteins = 0
fasta_entries = []

print(f"\nüöÄ Starting parallel download of {len(valid_df)} proteomes...\n")

with ThreadPoolExecutor(max_workers=10) as executor:  # You can try 20 or 30 depending on your network
    futures = {executor.submit(download_fasta_entry, row): row for _, row in valid_df.iterrows()}
    
    for future in as_completed(futures):
        fasta = future.result()
        if fasta:
            total_proteins += fasta.count(">")
            fasta_entries.append(fasta)

# === Save combined FASTA ===
with open(combined_fasta_path, "w") as f:
    f.writelines(fasta_entries)

print(f"\nüéØ Combined FASTA saved to: {combined_fasta_path}")
print(f"üß¨ Total proteins in database: {total_proteins}")
```

```python
from Bio import SeqIO
import pandas as pd
import re
from pathlib import Path

# === File paths ===
fasta_path = base_path / "Diamond_alignments" / "session_diamond_align_2025-05-15_20-37-25" / "diamond_15hits_bacteria.fasta"
metagenomics_path = base_path / "db_results_analysis" / "unique_taxa_in_metagendb.csv"

# === Step 1: Extract taxon names from FASTA (via OS=... in header)
fasta_taxa = set()
for record in SeqIO.parse(fasta_path, "fasta"):
    match = re.search(r"OS=([^=]+?)\s", record.description)
    if match:
        taxon_name = match.group(1).strip()
        fasta_taxa.add(taxon_name)

# === Step 2: Load taxa from metagenomics DB
df_meta = pd.read_csv(metagenomics_path)
df_meta.columns = df_meta.columns.str.strip()
meta_taxa = set(df_meta["taxon_name"].dropna().str.strip())

# === Step 3: Compare
overlap = fasta_taxa & meta_taxa
only_in_fasta = fasta_taxa - meta_taxa
only_in_meta = meta_taxa - fasta_taxa

# === Step 4: Report
total_fasta = len(fasta_taxa)
total_meta = len(meta_taxa)
overlap_count = len(overlap)
percent_overlap = (overlap_count / total_fasta) * 100 if total_fasta > 0 else 0

# === Step 5: Print
print(f"üß¨ Total unique taxa in DIAMOND 15-hits (bacteria): {total_fasta}")
print(f"üß¨ Total unique taxa in Metagenomics DB: {total_meta}")
print(f"‚úÖ Overlapping taxa: {overlap_count}")
print(f"‚ùå DIAMOND taxa not in Metagenomics: {len(only_in_fasta)}")
print(f"‚ùå Metagenomics taxa not in DIAMOND: {len(only_in_meta)}")
print(f"üìä Percent of DIAMOND taxa overlapping with Metagenomics: {percent_overlap:.2f}%")
```

```python
# Clustering the proteins in the database with MMseqs2
import subprocess
import os

# ‚úÖ Step 1: Define the input file path (Windows format)
input_file_path = r"C:\Users\Yusuf\OneDrive\LST\Derde_jaar\Y3Q4\Pipeline\Diamond_alignments\session_diamond_align_2025-05-15_20-37-25\diamond_aligned_from_organisms_15hits.fasta"

# ‚úÖ Step 2: Convert to WSL path
input_fasta = input_file_path.replace("\\", "/").replace("C:", "/mnt/c")
output_fasta = input_fasta.replace(".fasta", "_clustered90.fasta")

print(f"üîÅ Converted WSL input path:\n{input_fasta}")
print(f"üìÑ Output clustered FASTA will be:\n{output_fasta}")

# ‚úÖ Step 3: Define MMseqs2 commands
commands = [
    f'wsl mmseqs createdb {input_fasta} custom_db',
    'wsl mmseqs cluster custom_db clustered_db tmp --min-seq-id 0.9',
    'wsl mmseqs createsubdb clustered_db custom_db clustered_reps',
    f'wsl mmseqs result2flat custom_db custom_db clustered_reps {output_fasta}'
]

# ‚úÖ Step 4: Function to stream output in real-time
def run_with_streaming(cmd):
    print(f"\nüîÑ Running: {cmd}")
    process = subprocess.Popen(cmd, shell=True, stdout=subprocess.PIPE, stderr=subprocess.STDOUT, text=True)
    for line in process.stdout:
        print(line.strip())
    process.wait()
    if process.returncode != 0:
        raise subprocess.CalledProcessError(process.returncode, cmd)

# ‚úÖ Step 5: Run each MMseqs2 command with streaming
for cmd in commands:
    run_with_streaming(cmd)

print(f"\n‚úÖ Done! Clustered FASTA saved as:\n{output_fasta}")

# ‚úÖ Step 6: Clean up intermediate files (optional)
intermediate_files = [
    "custom_db", "custom_db_h", "custom_db.index", "custom_db.dbtype",
    "clustered_db", "clustered_db_h", "clustered_db.index", "clustered_db.dbtype",
    "clustered_db_seq", "clustered_db_seq_h", "clustered_db_seq.index", "clustered_db_seq.dbtype",
    "tmp"
]

for fname in intermediate_files:
    try:
        if os.path.isdir(fname):
            os.rmdir(fname)
        elif os.path.exists(fname):
            os.remove(fname)
        print(f"üßπ Deleted: {fname}")
    except Exception as e:
        print(f"‚ö†Ô∏è Could not delete {fname}: {e}")

print("‚úÖ Cleanup complete.")
```
